#  Indexing & Slicing

```python
# Indexing
a = np.diag(np.arange(5))
print(a)
print(a[1,1])
print(a[2,1])
```

```python
# Slicing
a = np.eye(5)
print(a)

print(a[2:4,:3]) #3rd and 4th lines, 3 first columns

a[:3,:3] = 4 # All elements specified by a slice can be easily modified
print(a)
```

```python
a = np.arange(10)
print(a)
a[2:9:3] # [start:end:step]
```
